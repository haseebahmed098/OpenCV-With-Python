```python
import cv2
from google.colab.patches import cv2_imshow

# Load image
image_path = "/content/guy.jpg"
image = cv2.imread(image_path)

# Check if image loaded successfully
if image is None:
    print("Error: Image not found!")
else:
    # Activity 3: Basic Image Operations

    ## 1. Resize Image
    new_width, new_height = 200, 300  # Set desired size
    resized_image = cv2.resize(image, (new_width, new_height))

    ## 2. Rotate Image (90 degrees clockwise)
    rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

    ## 3. Flip Image Horizontally
    flipped_image = cv2.flip(image, 1)

    # Activity 4: Apply Image Filtering Techniques

    ## 1. Apply Gaussian Blur
    kernel_size = 5  # Kernel size must be an odd number
    blurred_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

    ## 2. Convert Image to Grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Activity 5: Draw on Images to Annotate and Highlight Features

    ## 1. Draw a Rectangle
    start_x, start_y, end_x, end_y = 50, 50, 200, 200  # Define rectangle coordinates
    annotated_image = image.copy()  # Make a copy for annotation
    cv2.rectangle(annotated_image, (start_x, start_y), (end_x, end_y), (0, 255, 0), 2)

    ## 2. Add Text
    cv2.putText(annotated_image, 'Hello, OpenCV!', (50, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    # Activity 6: Save Processed Images
    cv2.imwrite('/mnt/data/resized_image.jpg', resized_image)
    cv2.imwrite('/mnt/data/rotated_image.jpg', rotated_image)
    cv2.imwrite('/mnt/data/flipped_image.jpg', flipped_image)
    cv2.imwrite('/mnt/data/blurred_image.jpg', blurred_image)
    cv2.imwrite('/mnt/data/gray_image.jpg', gray_image)
    cv2.imwrite('/mnt/data/annotated_image.jpg', annotated_image)

    # Display all processed images
    print("Original Image:")
    cv2_imshow(image)

    print("Resized Image:")
    cv2_imshow(resized_image)

    print("Rotated Image:")
    cv2_imshow(rotated_image)

    print("Flipped Image:")
    cv2_imshow(flipped_image)

    print("Blurred Image:")
    cv2_imshow(blurred_image)

    print("Grayscale Image:")
    cv2_imshow(gray_image)

    print("Annotated Image:")
    cv2_imshow(annotated_image)

    print("All processed images have been saved to /mnt/data/")
```

```python
import cv2
from google.colab.patches import cv2_imshow

# ✅ Step 1: Load the Image
image_path = "/content/balls.jpg"  # Change this if using a different image
image = cv2.imread(image_path)

# Check if the image is loaded successfully
if image is None:
    print("Error: Image not found!")
else:
    # --------------------------------------
    # LAB TASK 1: BASIC IMAGE OPERATIONS
    # --------------------------------------

    # ✅ Step 2: Resize the Image (Half of its original size)
    original_height, original_width = image.shape[:2]
    new_width, new_height = original_width // 2, original_height // 2
    resized_image = cv2.resize(image, (new_width, new_height))

    # ✅ Step 3: Rotate the Resized Image (180 degrees clockwise)
    rotated_image = cv2.rotate(resized_image, cv2.ROTATE_180)

    # ✅ Step 4: Flip the Rotated Image Vertically
    flipped_image = cv2.flip(rotated_image, 0)

    # --------------------------------------
    # LAB TASK 2: IMAGE FILTERING TECHNIQUES
    # --------------------------------------

    # ✅ Step 5: Apply Gaussian Blur (Kernel size: 5x5)
    blurred_image = cv2.GaussianBlur(flipped_image, (5, 5), 0)

    # ✅ Step 6: Convert the Blurred Image to Grayscale
    grayscale_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)

    # --------------------------------------
    # SAVE PROCESSED IMAGES
    # --------------------------------------
    cv2.imwrite('/mnt/data/resized_image.jpg', resized_image)
    cv2.imwrite('/mnt/data/rotated_image.jpg', rotated_image)
    cv2.imwrite('/mnt/data/flipped_image.jpg', flipped_image)
    cv2.imwrite('/mnt/data/blurred_image.jpg', blurred_image)
    cv2.imwrite('/mnt/data/grayscale_image.jpg', grayscale_image)

    # --------------------------------------
    # DISPLAY PROCESSED IMAGES
    # --------------------------------------
    print("Original Image:")
    cv2_imshow(image)

    print("Resized Image (Half of Original):")
    cv2_imshow(resized_image)

    print("Rotated Image (180 Degrees Clockwise):")
    cv2_imshow(rotated_image)

    print("Flipped Image (Vertically Flipped):")
    cv2_imshow(flipped_image)

    print("Blurred Image (Gaussian Blur 5x5):")
    cv2_imshow(blurred_image)

    print("Grayscale Image:")
    cv2_imshow(grayscale_image)

    print("All processed images have been saved to /mnt/data/")
```
