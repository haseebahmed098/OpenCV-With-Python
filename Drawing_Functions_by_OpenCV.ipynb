# **Task 1: Drawing Ellipses**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = np.zeros((500, 500, 3), dtype=np.uint8)  # Creating a black image

# Draw three ellipses
cv2.ellipse(image, (150, 150), (100, 50), 30, 0, 360, (0, 255, 0), 2)  # Green ellipse
cv2.ellipse(image, (300, 200), (80, 40), 0, 0, 360, (255, 0, 0), 2)  # Blue ellipse
cv2.ellipse(image, (250, 350), (60, 90), 45, 0, 360, (0, 0, 255), 2)  # Red ellipse

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

# **Task 2: Converting Ellipses to Polygons**

```python
# Convert an ellipse to a polygon
poly_points = cv2.ellipse2Poly((250, 350), (60, 90), 45, 0, 360, 10)

# Draw the polygon
cv2.polylines(image, [poly_points], isClosed=True, color=(255, 255, 0), thickness=2)

# Display the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

# **Task 3: Filling Convex Polygons**

```python
# Fill the convex polygon
cv2.fillConvexPoly(image, poly_points, (0, 255, 255))

# Display the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

# **Task 4: Drawing Rectangles**

```python
# Draw two rectangles
cv2.rectangle(image, (50, 50), (200, 150), (255, 0, 0), 3)  # Blue rectangle
cv2.rectangle(image, (250, 250), (400, 400), (0, 255, 0), -1)  # Green filled rectangle

# Display the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

# **Task 5: Drawing Lines and Arrowed Lines**

```python
# Draw a simple line
cv2.line(image, (50, 50), (400, 50), (255, 255, 255), 2)

# Draw an arrowed line
cv2.arrowedLine(image, (100, 400), (400, 300), (0, 255, 0), 3)

# Display the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

# **Task 6: Drawing Polylines**

```python
# Define points for the polyline
pts = np.array([[100, 100], [200, 50], [300, 100], [250, 200]], np.int32)
pts = pts.reshape((-1, 1, 2))

# Draw the polyline
cv2.polylines(image, [pts], isClosed=True, color=(255, 0, 255), thickness=2)

# Display the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

# **Task 7: Putting Text on Images**

```python
# Put text on the image
cv2.putText(image, 'Hello HASEEB AHMED!', (50, 45), cv2.FONT_HERSHEY_SIMPLEX, 5, (255, 255, 255), 10)

# Display the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```
